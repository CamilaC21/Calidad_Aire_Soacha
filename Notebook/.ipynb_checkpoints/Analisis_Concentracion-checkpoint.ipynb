```python
# =============================================================================
# 0. Importar libreríaspdf
# =============================================================================
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

from pathlib import Path
from shapely.geometry import Point, Polygon
import libpysal as ps
from esda import Moran, Moran_Local
from splot.esda import moran_scatterplot, lisa_cluster

# Para el diagrama de Voronoi:
from scipy.spatial import Voronoi

# =============================================================================
# 1. Cargar y preparar datos
# =============================================================================
# Rutas a archivos
root_folder = "/notebooks/"
MON_path = root_folder + "data_MAR_1.csv"              # Datos de monitoreo de sensores
excel_path = root_folder + "Coordenadas_Instituciones.xlsx"   # Coordenadas de instituciones
SHAPE_MUN = root_folder + "municipios_colombia.shp"         # Shapefile de municipios
SHAPE_REDVIAL = root_folder + "RedVial.shp"                 # Capa de red vial
SHAPE_SUELO = root_folder + "Usos_Suelo_Urbano_POT._Municipio_de_Soacha._2000_(capa).shp"  # Capa de suelo
SHAPE_COMUNAS = root_folder + "Comunas_Soacha_Magna.shp"    # << NUEVO: Capa de comunas

# Cargar shapefiles
gdf_mun = gpd.read_file(SHAPE_MUN)
gdf_road = gpd.read_file(SHAPE_REDVIAL)
gdf_suelo = gpd.read_file(SHAPE_SUELO)
gdf_comunas = gpd.read_file(SHAPE_COMUNAS)    # << NUEVO

# Cargar coordenadas desde Excel y CSV de monitoreo
df_coord = pd.read_excel(excel_path, sheet_name="Hoja1")
df_moni = pd.read_csv(MON_path)

# --- NUEVO: convertir 'date' a datetime y extraer día y hora ---
df_moni['date'] = pd.to_datetime(df_moni['date'])
df_moni['dia'] = df_moni['date'].dt.date
df_moni['hora'] = df_moni['date'].dt.hour

# Verifica el resultado
print(df_moni[['date', 'dia', 'hora']].head())

```

```python
df_moni.rename(columns={"station": "Estación"}, inplace=True)
```

```python
#-------------------------------------------------------------------------------------
# Fusionar por "Estación"
datos_combinados = pd.merge(df_coord, df_moni, on="Estación", how="inner")
print("Fusion inicial:")
print(datos_combinados.head())
```

```python
# Separar la columna 'Coordenadas' en 'Latitud' y 'Longitud'
datos_combinados[['Latitud', 'Longitud']] = datos_combinados['Coordenadas'].str.split(',', expand=True)
datos_combinados['Latitud'] = datos_combinados['Latitud'].astype(float)
datos_combinados['Longitud'] = datos_combinados['Longitud'].astype(float)
print("Tras separar coordenadas:")
print(datos_combinados.head())
```

```python
# Agregar estadísticas por estación (por ejemplo, promedio PM2.5)
station_summary = df_moni.groupby("Estación").agg({
    "PM25": ["mean", "std", "min", "max"],
    "Humidity": ["mean", "std", "min", "max"],
    "Temperature": ["mean", "std", "min", "max"]
}).reset_index()

station_summary.columns = [
    "Estación",
    "PM25_mean", "PM25_std", "PM25_min", "PM25_max",
    "Humidity_mean", "Humidity_std", "Humidity_min", "Humidity_max",
    "Temperature_mean", "Temperature_std", "Temperature_min", "Temperature_max"
]

print("Resumen estadístico por estación:")
print(station_summary.head())
```

```python
#Extraer coordenadas únicas por estación
coord_estaciones = datos_combinados[["Estación", "Latitud", "Longitud"]].drop_duplicates()
# Fusionar coord. con resumen
gdf_est = station_summary.merge(coord_estaciones, on="Estación")

# Crear geometría de puntos y GeoDataFrame
geometry = [Point(xy) for xy in zip(gdf_est["Longitud"], gdf_est["Latitud"])]
gdf_est = gpd.GeoDataFrame(gdf_est, geometry=geometry, crs="EPSG:4326")

# Reproyectar al CRS EPSG:9377 (MAGNA-SIRGAS para zona de Soacha)
gdf_est = gdf_est.to_crs(epsg=9377)
gdf_mun = gdf_mun.to_crs(epsg=9377)
gdf_road = gdf_road.to_crs(epsg=9377)
gdf_suelo = gdf_suelo.to_crs(epsg=9377)
gdf_comunas = gdf_comunas.to_crs(epsg=9377)   # << NUEVO


# Filtrar el municipio de Soacha (ajusta el nombre según corresponda)
gdf_soacha = gdf_mun[gdf_mun["MPIO_CNMBR"].str.upper() == "SOACHA"]
```

```python
print("Comunas de Soacha (primeros registros):")
print(gdf_comunas.head())
print("CRS comunas:", gdf_comunas.crs)
```

```python
# Asumimos que gdf_est y gdf_comunas ya están en EPSG:9377

# Realizar join espacial: añade los atributos de comuna a cada estación según su ubicación
gdf_est_comuna = gpd.sjoin(gdf_est, gdf_comunas, how='left', predicate='within')

# Verifica el resultado
print("Estaciones con comuna asignada:")
print(gdf_est_comuna[['Estación', 'PM25_mean', 'geometry', 'COMUNA']].head())
```

```python
# Calcula el promedio de PM2.5 de las estaciones dentro de cada comuna
pm25_por_comuna = gdf_est_comuna.groupby('COMUNA')['PM25_mean'].mean().reset_index()

# Renombra columna para claridad
pm25_por_comuna.rename(columns={'PM25_mean': 'PM25_comuna_mean'}, inplace=True)

print("Promedio de PM2.5 por comuna:")
print(pm25_por_comuna)
```

```python
# Unir el promedio a la capa de comunas para visualizar en un mapa
gdf_comunas_pm25 = gdf_comunas.merge(pm25_por_comuna, left_on='COMUNA', right_on='COMUNA', how='left')

# Verifica
print(gdf_comunas_pm25[['COMUNA', 'PM25_comuna_mean']])
```

```python
def clasificar_pm25(valor):
    if valor <= 15:
        return "Verde"
    elif valor <= 25:
        return "Amarillo"
    elif valor <= 37:
        return "Naranja"
    else:
        return "Rojo intenso"
    
color_map = {
    "Verde": "#43a047",
    "Amarillo": "#fbc02d",
    "Naranja": "#fb8c00",
    "Rojo intenso": "#d32f2f"
}
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
from scipy.spatial import cKDTree
```

```python
Coord_estaciones = datos_combinados[["Estación", "Latitud", "Longitud"]].drop_duplicates()
# Fusionar coord. con resumen
gdf_est = station_summary.merge(coord_estaciones, on="Estación")

# Crear geometría de puntos y GeoDataFrame
geometry = [Point(xy) for xy in zip(gdf_est["Longitud"], gdf_est["Latitud"])]
gdf_est = gpd.GeoDataFrame(gdf_est, geometry=geometry, crs="EPSG:4326")

# Reproyectar al CRS EPSG:9377 (MAGNA-SIRGAS para zona de Soacha)
gdf_est = gdf_est.to_crs(epsg=9377)
gdf_mun = gdf_mun.to_crs(epsg=9377)
gdf_road = gdf_road.to_crs(epsg=9377)
gdf_suelo = gdf_suelo.to_crs(epsg=9377)
gdf_comunas = gdf_comunas.to_crs(epsg=9377)   # << NUEVO


# Filtrar el municipio de Soacha (ajusta el nombre según corresponda)
gdf_soacha = gdf_mun[gdf_mun["MPIO_CNMBR"].str.upper() == "SOACHA"]
```

```python
def idw_interpolation(xy, values, xi, yi, power=2):
    tree = cKDTree(xy)
    grid = np.c_[xi.ravel(), yi.ravel()]
    dist, idx = tree.query(grid, k=len(xy))
    weights = 1 / (dist**power)
    weights[dist == 0] = 1e12
    weighted_values = np.sum(weights * values[idx], axis=1) / np.sum(weights, axis=1)
    return weighted_values.reshape(xi.shape)
```

```python
minx, miny, maxx, maxy = gdf_comunas.total_bounds
grid_x, grid_y = np.mgrid[minx:maxx:200j, miny:maxy:200j]
```

```python
# Promedio total por estación
pm25_promedio = df_moni.groupby('station')['PM25'].mean()
gdf_est['PM25_mean'] = gdf_est['Estación'].map(pm25_promedio)

xy = np.array([(geom.x, geom.y) for geom in gdf_est.geometry])
values = gdf_est['PM25_mean'].values
pm25_grid = idw_interpolation(xy, values, grid_x, grid_y)

# --- Graficar ---
fig, ax = plt.subplots(figsize=(10, 8))
gdf_suelo.plot(ax=ax, color='lightgrey', linewidth=0.3, edgecolor='white', alpha=0.6)
gdf_comunas.boundary.plot(ax=ax, color='black', linewidth=0.7)
im = ax.imshow(
    pm25_grid.T, extent=(minx, maxx, miny, maxy),
    origin='lower', cmap='hot', alpha=0.6
)
gdf_est.plot(ax=ax, color='blue', marker='o', markersize=35, label='Estación')
plt.colorbar(im, ax=ax, label='PM2.5 (µg/m³)')
plt.title('Mapa de calor PM2.5 - Promedio total')
plt.legend()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter


# Cargar shapefiles
gdf_mun = gpd.read_file(SHAPE_MUN)
gdf_road = gpd.read_file(SHAPE_REDVIAL)
gdf_suelo = gpd.read_file(SHAPE_SUELO)
gdf_comunas = gpd.read_file(SHAPE_COMUNAS)    # << NUEVO


# Reproyectar al CRS EPSG:9377 (MAGNA-SIRGAS para zona de Soacha)
gdf_est = gdf_est.to_crs(epsg=9377)
gdf_mun = gdf_mun.to_crs(epsg=9377)
gdf_road = gdf_road.to_crs(epsg=9377)
gdf_suelo = gdf_suelo.to_crs(epsg=9377)
gdf_comunas = gdf_comunas.to_crs(epsg=9377)   # << NUEVO


xy = np.vstack([gdf_est.geometry.x, gdf_est.geometry.y]).T
minx, miny, maxx, maxy = gdf_comunas.total_bounds
grid_x = np.linspace(minx, maxx, 200)
grid_y = np.linspace(miny, maxy, 200)
grid_x, grid_y = np.meshgrid(grid_x, grid_y)

# —————————————————————————
# 2. Funciones de animación

fig, ax = plt.subplots(figsize=(10, 8))

def init():
    ax.clear()
    gdf_suelo.plot(ax=ax, color='lightgrey', linewidth=0.3, edgecolor='white', alpha=0.6)
    gdf_comunas.boundary.plot(ax=ax, color='black', linewidth=0.7)
    ax.set_axis_off()
    return []

def update(hora):
    ax.clear()
    # Fondo y comunas
    gdf_suelo.plot(ax=ax, color='lightgrey', linewidth=0.3, edgecolor='white', alpha=0.6)
    gdf_comunas.boundary.plot(ax=ax, color='black', linewidth=0.7)

    # PM2.5 promedio por estación en esta hora
    pm25_hora = (
        df_moni[df_moni['hora'] == hora]
        .groupby('station')['PM25']
        .mean()
    )
    gdf_est['PM25_hora'] = gdf_est['Estación'].map(pm25_hora).fillna(0)

    # IDW simple (puedes reemplazar por tu propia función)
    dist = np.hypot(
        grid_x[...,None] - xy[:,0],
        grid_y[...,None] - xy[:,1]
    )
    dist[dist==0] = 1e-10
    weights = 1 / dist**2
    pm25_grid = (weights * gdf_est['PM25_hora'].values).sum(axis=2) / weights.sum(axis=2)

    # Mostrar raster y estaciones
    im = ax.imshow(
        pm25_grid.T,
        extent=(minx, maxx, miny, maxy),
        origin='lower',
        cmap='hot',
        alpha=0.6
    )
    gdf_est.plot(ax=ax, color='blue', marker='o', markersize=35, label='Estación')
    ax.set_title(f'PM₂.₅ Promedio Hora {hora:02d}:00', fontsize=14)
    ax.legend(loc='lower right')
    ax.set_axis_off()
    return [im]

# —————————————————————————
# 3. Crear y guardar GIF con PillowWriter

ani = FuncAnimation(
    fig, update,
    frames=range(24),
    init_func=init,
    blit=False,
    repeat=True
)

writer = PillowWriter(fps=1)  # un frame por segundo
ani.save('PM25_por_hora.gif', writer=writer)

print("✅ GIF generado: PM25_por_hora.gif")
```

```python
import os
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import imageio.v2 as imageio
import shutil

# === 1. Cargar datos ===
# Ajusta las rutas de tus archivos:
SHAPE_MUN = "/notebooks/municipios_colombia.shp"
SHAPE_REDVIAL = "/notebooks/RedVial.shp"
SHAPE_SUELO = "/notebooks/ClasificacionSuelo.shp"
SHAPE_COMUNAS = "/notebooks/Comunas_Soacha_Magna.shp"
MON_path = "/notebooks/data_MAR_1.csv"
# Si aún no tienes gdf_est: crea el GeoDataFrame de estaciones
excel_path = "/notebooks/Coordenadas_Instituciones.xlsx"
df_coord = pd.read_excel(excel_path, sheet_name="Hoja1")
df_moni = pd.read_csv(MON_path)
df_moni['date'] = pd.to_datetime(df_moni['date'])
df_moni['hora'] = df_moni['date'].dt.hour

# --- crea gdf_est si no existe (ajusta columnas según tu archivo Excel) ---
df_coord[['Latitud', 'Longitud']] = df_coord['Coordenadas'].str.split(',', expand=True)
df_coord['Latitud'] = df_coord['Latitud'].astype(float)
df_coord['Longitud'] = df_coord['Longitud'].astype(float)
geometry = gpd.points_from_xy(df_coord['Longitud'], df_coord['Latitud'])
gdf_est = gpd.GeoDataFrame(df_coord[['Estación']], geometry=geometry, crs="EPSG:4326")

# Cargar shapefiles
gdf_mun = gpd.read_file(SHAPE_MUN)
gdf_road = gpd.read_file(SHAPE_REDVIAL)
gdf_suelo = gpd.read_file(SHAPE_SUELO)
gdf_comunas = gpd.read_file(SHAPE_COMUNAS)

# Reproyectar todo a EPSG:9377
gdf_est = gdf_est.to_crs(epsg=9377)
gdf_mun = gdf_mun.to_crs(epsg=9377)
gdf_road = gdf_road.to_crs(epsg=9377)
gdf_suelo = gdf_suelo.to_crs(epsg=9377)
gdf_comunas = gdf_comunas.to_crs(epsg=9377)

# Grilla para interpolación
xy = np.vstack([gdf_est.geometry.x, gdf_est.geometry.y]).T
minx, miny, maxx, maxy = gdf_comunas.total_bounds
grid_x = np.linspace(minx, maxx, 200)
grid_y = np.linspace(miny, maxy, 200)
grid_x, grid_y = np.meshgrid(grid_x, grid_y)

# Colores y norma
cmap_norma = mcolors.ListedColormap(['#43a047', '#fbc02d', '#fb8c00', '#d32f2f'])
bounds = [0, 15, 25, 37, 1000]
norm = mcolors.BoundaryNorm(bounds, cmap_norma.N)
leyenda = [
    mpatches.Patch(color='#43a047', label='≤15 µg/m³ (Verde)'),
    mpatches.Patch(color='#fbc02d', label='>15-25 µg/m³ (Amarillo)'),
    mpatches.Patch(color='#fb8c00', label='>25-37 µg/m³ (Naranja)'),
    mpatches.Patch(color='#d32f2f', label='>37 µg/m³ (Rojo intenso)')
]

# === 2. Función para dibujar cada frame ===
def plot_frame(ax, hora):
    ax.clear()
    gdf_suelo.plot(ax=ax, color='lightgrey', linewidth=0.3, edgecolor='white', alpha=0.6)
    gdf_comunas.boundary.plot(ax=ax, color='black', linewidth=0.7)
    pm25_hora = (
        df_moni[df_moni['hora'] == hora]
        .groupby('station')['PM25']
        .mean()
    )
    gdf_est['PM25_hora'] = gdf_est['Estación'].map(pm25_hora).fillna(0)
    # --- IDW interpolación simple ---
    dist = np.hypot(
        grid_x[...,None] - xy[:,0],
        grid_y[...,None] - xy[:,1]
    )
    dist[dist==0] = 1e-10
    weights = 1 / dist**2
    pm25_grid = (weights * gdf_est['PM25_hora'].values).sum(axis=2) / weights.sum(axis=2)
    im = ax.imshow(
        pm25_grid.T,
        extent=(minx, maxx, miny, maxy),
        origin='lower',
        cmap=cmap_norma,
        norm=norm,
        alpha=0.7
    )
    gdf_est.plot(ax=ax, color='blue', marker='o', markersize=35, label='Estación')
    ax.set_title(f'PM₂.₅ Promedio Hora {hora:02d}:00', fontsize=16)
    ax.text(
        0.97, 0.90, f'{hora:02d}:00',
        fontsize=32, color='black', ha='right', va='top',
        transform=ax.transAxes, bbox=dict(facecolor='white', alpha=0.6, edgecolor='none')
    )
    legend1 = ax.legend(handles=leyenda, loc='lower left', fontsize=11, title='Norma Resolución 2254/2017', title_fontsize=12)
    ax.add_artist(legend1)
    cbar = plt.colorbar(im, ax=ax, label='PM2.5 (µg/m³)', ticks=[0, 15, 25, 37, 50])
    cbar.ax.tick_params(labelsize=11)
    ax.set_axis_off()

# === 3. Generar imágenes por hora ===
output_dir = "gif_horas_pm25"
os.makedirs(output_dir, exist_ok=True)
filepaths = []

for hora in range(24):
    fig, ax = plt.subplots(figsize=(10, 8))
    plot_frame(ax, hora)
    filename = os.path.join(output_dir, f"PM25_hora_{hora:02d}.png")
    plt.savefig(filename, dpi=120, bbox_inches='tight')
    filepaths.append(filename)
    plt.close(fig)

# === 4. Crear GIF animado ===
images = [imageio.imread(fp) for fp in filepaths]
imageio.mimsave('PM25_por_hora.gif', images, duration=3.0)  # 3 seg/frame

print("✅ GIF generado: PM25_por_hora.gif")

# === 5. Opcional: borrar imágenes temporales ===
shutil.rmtree(output_dir)
```
