# Analisi  estadistico No 2 an√°lisis LISA (Local Indicators of Spatial Association)

Este fragmento de c√≥digo en Python genera un mapa que visualiza la ubicaci√≥n de las estaciones de monitoreo sobre el municipio de Soacha, destacando aquellas clasificadas como High‚ÄìHigh (HH) seg√∫n el an√°lisis LISA.

Grafica todas las estaciones con puntos en gris claro (lightgray) y tama√±o medio. Estas son tus estaciones de referencia.

Dibuja las estaciones clasificadas como High‚ÄìHigh (HH) y estad√≠sticamente significativas (p < 0.05) en color rojo carmes√≠ (crimson) y con un tama√±o ligeramente mayor.
üî¥ Estos son tus focos locales de alta contaminaci√≥n (es decir, estaciones con PM2.5 alto rodeadas por otras tambi√©n altas).

En el contexto del an√°lisis LISA (Local Indicators of Spatial Association), ‚ÄúHigh‚ÄìHigh‚Äù (o ‚ÄúHH‚Äù) se refiere a una estaci√≥n o punto que:
- Tiene un valor alto de la variable que est√°s analizando (en tu caso, promedio de PM2.5),
- Y est√° rodeada por vecinos que tambi√©n tienen valores altos de esa misma variable.
üå°Ô∏è En otras palabras: es un hotspot estad√≠sticamente significativo, donde se concentra contaminaci√≥n elevada y adem√°s hay ‚Äúefecto barrio‚Äù: las estaciones cercanas tambi√©n reportan niveles altos.
Esto se interpreta as√≠:
- High‚ÄìHigh (HH): foco local de contaminaci√≥n ‚Üí podr√≠a relacionarse con emisiones industriales, tr√°fico o quema cerca.
- Low‚ÄìLow (LL): punto limpio rodeado de puntos limpios ‚Üí ‚Äúcold-spot‚Äù.
- High‚ÄìLow (HL) y Low‚ÄìHigh (LH): puntos que son outliers respecto a su entorno.
Y solo se considera confiable cuando el p-valor local < 0.05 (es decir, el agrupamiento no es producto del azar).

```python
# Librer√≠as para an√°lisis de datos
import geopandas as gpd
import pandas as pd
import numpy as np
import contextily as cx
from pathlib import Path
from shapely.geometry import Point

# Visualizaci√≥n
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Estad√≠stica temporal y correlaci√≥n
import statsmodels.tsa.stattools as ts
from scipy.stats import pearsonr

# An√°lisis espacial
import geopandas as gpd
from shapely.geometry import Point
import libpysal as lp
from esda import Moran, Moran_Local

# Rutas
root_folder = "/notebooks/"
MON_path = root_folder + "data_MAR_1.csv"
excel_path = root_folder + "Coordenadas_Instituciones.xlsx"
SHAPE_MUN = root_folder + "municipios_colombia.shp"
SHAPE_REDVIAL = root_folder + "RedVial.shp"
SHAPE_SUELO = root_folder + "Usos_Suelo_Urbano_POT._Municipio_de_Soacha._2000_(capa).shp"

# Carga de archivos
gdf_mun = gpd.read_file(SHAPE_MUN)
gdf_road = gpd.read_file(SHAPE_REDVIAL)
gdf_suelo = gpd.read_file(SHAPE_SUELO)

# Cargar coordenadas desde Excel
df_coord = pd.read_excel(excel_path, sheet_name="Hoja1")

# Cargar CSV de datos de monitoreo
df_moni = pd.read_csv(MON_path)

# Mostrar columnas para inspecci√≥n r√°pida
print("Municipios:", gdf_mun.columns)
print("Coordenadas:", df_coord.columns)
print("Red vial:", gdf_road.columns)
print("Usos del suelo:", gdf_suelo.columns)
print("Datos de monitoreo:", df_moni.columns)
```

```python
# en esta inea  se almacenan coordenadas instituciones en un dateframe
df_moni.rename(columns={"station": "Estaci√≥n"}, inplace=True)

# Fusionar las tablas en base a la columna com√∫n "Instituci√≥n"
datos_combinados = pd.merge(df_coord, df_moni, on="Estaci√≥n", how="inner")
# Mostrar las primeras filas para verificar la fusi√≥n
print("Fusion inicial:", datos_combinados.head()) # print(datos_combinados.head())
```

```python
# Separar la columna 'Coordenadas' en columnas num√©ricas 'Latitud' y 'Longitud'
datos_combinados[['Latitud', 'Longitud']] = datos_combinados['Coordenadas'].str.split(',', expand=True)

# Limpiar espacios en blanco y convertir los valores a tipo float
datos_combinados['Latitud'] = datos_combinados['Latitud'].astype(float)
datos_combinados['Longitud'] = datos_combinados['Longitud'].astype(float)
```

```python
print(datos_combinados.head())
print("Tras separar coordenadas:", datos_combinados.head())
print("Columnas disponibles:", list(datos_combinados.columns))
```

```python
# Agrupar por estaci√≥n y calcular estad√≠sticas de PM2.5, Humedad y Temperatura
station_summary = df_moni.groupby("Estaci√≥n").agg({
    "PM25": ["mean", "std", "min", "max"],
    "Humidity": ["mean", "std", "min", "max"],
    "Temperature": ["mean", "std", "min", "max"]
}).reset_index()

# Aplanar nombres de columnas del MultiIndex
station_summary.columns = [
    "station",
    "PM25_mean", "PM25_std", "PM25_min", "PM25_max",
    "Humidity_mean", "Humidity_std", "Humidity_min", "Humidity_max",
    "Temperature_mean", "Temperature_std", "Temperature_min", "Temperature_max"
]
```

```python
# Mostrar la tabla resultante
print("\nResumen estad√≠stico por estaci√≥n:")
print(station_summary.head(10))
```

```python
# Crear geometr√≠a de puntos a partir de latitud y longitud
geometry = [Point(xy) for xy in zip(datos_combinados["Longitud"], datos_combinados["Latitud"])]
gdf_estaciones = gpd.GeoDataFrame(datos_combinados, geometry=geometry, crs="EPSG:4326")

# Reproyectar al sistema de coordenadas del shapefile (Soacha probablemente est√© en EPSG:3116 o 3857)
gdf_estaciones = gdf_estaciones.to_crs(gdf_mun.crs)
```

```python
# Boxplot de PM2.5 por estaci√≥n
plt.figure(figsize=(12, 6))
sns.boxplot(data=datos_combinados, x="Estaci√≥n", y="PM25", palette="OrRd")
plt.title("Distribuci√≥n de PM2.5 por estaci√≥n")
plt.xlabel("Estaci√≥n")
plt.ylabel("PM2.5 (Œºg/m¬≥)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Visualizar distribuci√≥n con boxplots
Esto nos permite comparar la variabilidad de PM2.5 entre estaciones, no solo su promedio.
- Cada caja representa el rango intercuartil (Q1 a Q3) de valores.
- La l√≠nea en medio es la mediana.
- Los puntos fuera de la caja son outliers.
- Si una estaci√≥n tiene una caja m√°s alta y sin superposici√≥n con otras, indica mayor nivel y posible hot‚Äêspot.

```python
# Seleccionar coordenadas √∫nicas por estaci√≥n
coord_estaciones = datos_combinados[["Estaci√≥n", "Latitud", "Longitud"]].drop_duplicates()

# Fusionar con resumen estad√≠stico (ya tiene 13 estaciones)
gdf_est = station_summary.merge(coord_estaciones, left_on="Estaci√≥n", right_on="Estaci√≥n")

# Crear geometr√≠a y GeoDataFrame
geometry = [Point(xy) for xy in zip(gdf_est["Longitud"], gdf_est["Latitud"])]
gdf_est = gpd.GeoDataFrame(gdf_est, geometry=geometry, crs="EPSG:4326").to_crs(epsg=3857)
# Convertir al sistema de referencia EPSG:9377
gdf_est = gdf_est.to_crs(epsg=9377)
```

```python
import libpysal as ps
from esda import Moran
from splot.esda import moran_scatterplot

# Extraer coordenadas
coords = np.array(list(zip(gdf_est.geometry.x, gdf_est.geometry.y)))

# Crear matriz de 4 vecinos m√°s cercanos
knn = ps.weights.KNN.from_array(coords, k=4)
knn.transform = "R"  # normalizar por fila
```

```python
# √çndice de Moran para PM2.5_mean
moran_pm = Moran(gdf_est["PM25_mean"], knn)

print(f"Moran‚Äôs I: {moran_pm.I:.4f}")
print(f"p-valor (normalidad): {moran_pm.p_norm:.4f}")
```

```python
fig, ax = moran_scatterplot(moran_pm, aspect_equal=True)
ax.set_title("Diagrama de Moran para PM2.5 promedio por estaci√≥n")
plt.tight_layout()
plt.show()
```

# Interpretaci√≥n Indice de MORAN
- Moran‚Äôs I > 0 y p < 0.05: agrupamiento de estaciones con valores similares ‚Üí autocorrelaci√≥n significativa ‚Üí PM2.5 no es aleatorio ‚Üí podr√≠as justificar an√°lisis espacial.
- Moran‚Äôs I ‚âà 0 o p > 0.05: no hay agrupamiento espacial ‚Üí distribuci√≥n aleatoria.
- El gr√°fico divide tus estaciones en cuadrantes:
- Q1 (HH): valores altos rodeados de valores altos.
- Q3 (LL): valores bajos entre bajos.
- Q2/Q4: posibles outliers espaciales.


Primero, asegur√©monos de que el GeoDataFrame gdf_est contiene la media de PM2.5 por estaci√≥n y est√© en EPSG:9377 (ya lo confirmamos)

```python
from esda import Moran_Local
from splot.esda import lisa_cluster
import matplotlib.pyplot as plt

# Calcular LISA para la variable PM2.5_media
lisa = Moran_Local(station_summary["PM25_mean"], knn)  # Usamos la misma matriz de pesos knn

# Agregar resultados al GeoDataFrame
gdf_est["lisa_q"] = lisa.q  # Cuadrante (1‚Äì4)
gdf_est["lisa_p"] = lisa.p_sim  # p-valor local

# Clasificaciones interpretables
gdf_est["lisa_type"] = gdf_est["lisa_q"].map({
    1: "High-High",   # Hotspot
    2: "Low-High",    # Outlier
    3: "Low-Low",     # Coldspot
    4: "High-Low"     # Outlier inverso
})
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
lisa_cluster(lisa, gdf_est, p=0.05, ax=ax, legend=True)
gdf_soacha.boundary.plot(ax=ax, color="black", linewidth=1)
ax.set_title("Clusters LISA (PM2.5 promedio por estaci√≥n)")
plt.tight_layout()
plt.show()
```

# Evaluaci√≥n para usar LUR puro
Si menos del 10‚Äì15% de tus estaciones muestran agrupamientos significativos, LUR puro puede ser adecuado.
Si hay m√∫ltiples puntos HH o HL con p < 0.05, considera:
- LUR con t√©rminos espaciales (Spatial Lag / Error)
- LUR local (GWR)

```python
# ¬øCu√°ntas estaciones tienen p < 0.05?
signif_pct = (gdf_est["lisa_p"] < 0.05).mean()
print(f"Porcentaje de estaciones en clusters significativos: {signif_pct:.2%}")
```

```python
# Filtrar estaciones con etiqueta HH (lisa_q == 1) y p-valor < 0.05
hotspots = gdf_est[(gdf_est["lisa_q"] == 1) & (gdf_est["lisa_p"] < 0.05)]

# Mostrar tabla resumen
print(f"üî¥ N√∫mero de estaciones en High‚ÄìHigh cl√∫ster significativo: {len(hotspots)}")
display(hotspots[["Estaci√≥n", "PM25_mean", "lisa_p"]])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
gdf_soacha.boundary.plot(ax=ax, color="black")
gdf_est.plot(ax=ax, color="lightgray", markersize=80, label="Todas las estaciones")
hotspots.plot(ax=ax, color="crimson", markersize=100, label="Hotspots High‚ÄìHigh")

for x, y, label in zip(hotspots.geometry.x, hotspots.geometry.y, hotspots["Estaci√≥n"]):
    ax.text(x, y, label, fontsize=9, ha="right")

ax.set_title("Estaciones clasificadas como High‚ÄìHigh (LISA)")
ax.legend()
plt.axis("off")
plt.tight_layout()
plt.show()
```

# Podemos usar un modelo LUR puro?
S√≠, y de hecho estos resultados lo respaldan.

# 1. Bajo agrupamiento espacial (p > 0.05)
No hay evidencia estad√≠sticamente significativa de que los valores de PM2.5 est√©n agrupados en el espacio. Es decir, no est√°n "pegados entre s√≠" como para requerir un t√©rmino espacial.
# 2. No hay cl√∫steres locales significativos (High‚ÄìHigh)
Eso descarta focos dominantes que pudieran sesgar los coeficientes de un modelo lineal tradicional.
# 3. Estaciones con independencia espacial
Dado que no hay dependencia espacial global ni local, tus estaciones pueden tratarse como unidades independientes, y eso es justamente el supuesto principal de un modelo LUR.

El modelo LUR (Land Use Regression) asume que la variabilidad en PM2.5 puede explicarse con variables como:
- Uso de suelo
- Densidad vial
- Cobertura vegetal
- Temperatura / Humedad ...sin necesidad de incluir estructura espacial expl√≠cita.
