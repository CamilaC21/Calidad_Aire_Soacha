# **Análisis de vulnerabilidad por contaminación atmosférica**   
En este notebook se realiza el análisis de vulnerabilidad por contaminación atmosferica, teniendo como insumo el uso de suelos y la ubicación de los centros medicos.   
Esta clasificación es de baja, moderada y alta.

Inicialmente se importan las librerias

```python
import os
import rasterio
import numpy as np
import pandas as pd
import geopandas as gpd
from rasterio.transform import from_bounds
from rasterio.features import rasterize
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

**Preparación de datos**   
En la siguiente celda se lleva a cabo el cargue y preparación de la información a necesitar en el transcurso del proyecto.

```python
root_folder=r"/notebooks/Proyecto/Datos"

# Leer todos los datos
Suelo = root_folder+r"/UsoSuelo.shp"
Municipio = root_folder+r"/Soacha.shp"
Medico = root_folder+r"/CentrosMedicos.shp"

# Leer el archivo shapefile del uso del suelo
UsoSuelo = gpd.read_file(Suelo)
UsoSuelo = UsoSuelo.to_crs(epsg=9377)
print("CRS Uso Suelo:", UsoSuelo.crs)
print(UsoSuelo["USO_SUELO"])

# Leer el archivo shapefile del municipio de Soacha
Soacha = gpd.read_file(Municipio)
Soacha = Soacha.to_crs(epsg=9377)
print("CRS Municipio de Soacha:", Soacha.crs)

# Leer el archivo shapefile del centros medicos
CentrosSalud = gpd.read_file(Medico)
CentrosSalud = CentrosSalud.to_crs(epsg=9377)
print("CRS Centros de Salud:", CentrosSalud.crs)
print(CentrosSalud.head(1))
```

**Reclasificar usos del suelo por vulnerabilidad**   
En la siguiente celda se lleva a cabo la reclasificación para cada uso del suelo de acuerdo a lo siguiente:   
* Zonas Agricolas o rurales ---> Baja (1)
* Zonas Institucionales o sociales ---> Alta (3)
* Parque actividad economica ---> Alta (3)
* Planta de tratamiento de aguas residuales ---> Moderada (2)
* Parque metropolitano ---> Alta (3)
* Protección o preservación ambiental ---> Baja (1)
* Residencial ---> Alta (3)
* Área de actividad multiple ---> Moderada (2)

```python
# Establecer la reclasificación
Reclasificacion = {"ZONA AGRICOLA SOSTENIBLE": 1, 
                   "ZONA INSTITUCIONAL SERVICIOS": 3,
                   "ZONA INSTITUCIONAL 2 DE SERVICIOS ISS": 3,
                   "PARQUE ACTIVIDAD ECONOMICA": 3,
                   "PLANTA DE TRATAMIENTO DE AGUAS RESIDUALES": 2,
                   "UNIDADES INTEGRALES DE DESARROLLO SOCIAL": 3,
                   "PARQUE METROPOLITANO": 3,
                   "PROTECCION AMBIENTAL": 1,
                   "RESIDENCIAL": 3,
                   "PRESERVACION AMBIENTAL": 1,
                   "AREA DE ACTIVIDAD MULTIPLE": 2}
```

```python
# Crear columna con clasificación 
UsoSuelo["Clasificación"] = UsoSuelo["USO_SUELO"].map(Reclasificacion)
print(UsoSuelo[["USO_SUELO", "Clasificación"]])
```

```python
# Rasterizar la capa de uso del suelo
# Crear la malla para rasterizar
Extension = Soacha.total_bounds
res = 100
Ancho = int((Extension[2] - Extension[0])/ res)
Alto = int((Extension[3] - Extension[1])/ res)
transform = from_bounds(*Extension, Ancho, Alto)

# Rasterizar
shapes = ((geom, value) for geom, value in zip(UsoSuelo.geometry, UsoSuelo["Clasificación"]))
Raster = rasterize(shapes=shapes, out_shape=(Alto, Ancho), transform=transform, fill=0, dtype="uint8")

# Guardar el raster clasificado
with rasterio.open(
    "/notebooks/Proyecto/Calidad_Aire_Soacha/Resultados/Raster/Reclasificados/UsoSueloVul_reclass.tif",
    "w",
    driver = "GTiff",
    height = Alto,
    width = Ancho,
    count = 1,
    dtype = "uint8",
    crs = UsoSuelo.crs,
    transform = transform,
)as dst:
    dst.write(Raster, 1)
```

```python
with rasterio.open("/notebooks/Proyecto/Calidad_Aire_Soacha/Resultados/Raster/Reclasificados/UsoSueloVul_reclass.tif") as src:
    UsoSueloVul_reclass = src.read(1)
    transform = src.transform

colores = ["green", "yellow", "red"]
cmap = ListedColormap(colores)

plt.figure(figsize=(8, 10))
plt.imshow(UsoSueloVul_reclass, cmap=cmap, extent=(transform[2], transform[2] + transform[0]*UsoSueloVul_reclass.shape[1],
                                            transform[5] + transform[4]*UsoSueloVul_reclass.shape[0], transform[5]))


cbar = plt.colorbar(ticks=[1, 2, 3])
cbar.ax.set_yticklabels(["Baja", "Moderada", "Alta"])
cbar.set_label("Categoría de vulnerabilidad")

plt.title("Vulnerabilidad de PM2.5 de acuerdo al uso del suelo")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(False)
plt.savefig(f"/notebooks/Proyecto/Calidad_Aire_Soacha/Resultados/Vulnerabilidad_PM25.png", dpi=300)
plt.show()
```

**Reclasificar zonas de centros medicos por vulnerabilidad**   
En la siguiente celda se lleva a cabo la reclasificación para las áreas en donde hay presencia de centros medicos, de acuerdo a lo siguiente:   
* Zonas donde no hay centros medicos ---> Baja (1)
* Zonas donde hay centros medicos ---> Alta (3)

```python
# Crear un buffer de 150 metros para las áreas de influencia de los centros medicos
Buffer = CentrosSalud.buffer(150)
BufferCM = gpd.GeoDataFrame(geometry=Buffer, crs=Soacha.crs)
BufferCM["Vulnerabilidad"] = 3

# Crear el resto del área de Soacha con vulnerabilidad baja (1)
Vul_Baja = gpd.GeoDataFrame(geometry=[Soacha.geometry.union_all()], crs=Soacha.crs)
Vul_Baja["Vulnerabilidad"] = 1
Vul_Baja["geometry"] = Vul_Baja.difference(BufferCM.geometry.union_all())
Vulnerabilidad = gpd.GeoDataFrame(pd.concat([BufferCM, Vul_Baja], ignore_index=True), crs=Soacha.crs)

# Crear la malla para rasterizar
Extension = Soacha.total_bounds
res = 100
Ancho = int((Extension[2] - Extension[0])/ res)
Alto = int((Extension[3] - Extension[1])/ res)
transform = from_bounds(*Extension, Ancho, Alto)

# Rasterizar
shapes = ((geom, val) for geom, val in zip(Vulnerabilidad.geometry, Vulnerabilidad["Vulnerabilidad"]))
raster = rasterize(shapes=shapes, out_shape=(Alto, Ancho), transform=transform, fill=0, dtype="uint8")

# Guardas el raster
with rasterio.open(
    "/notebooks/Proyecto/Calidad_Aire_Soacha/Resultados/Raster/Reclasificados/CentroSldVul_reclass.tif",
    "w",
    driver = "GTiff",
    height = Alto,
    width = Ancho,
    count = 1,
    dtype = "uint8",
    crs = Vulnerabilidad.crs,
    transform = transform,
)as dst:
    dst.write(raster, 1)
```

```python
with rasterio.open("/notebooks/Proyecto/Calidad_Aire_Soacha/Resultados/Raster/Reclasificados/CentroSldVul_reclass.tif") as src:
    CentroSldVul_reclass = src.read(1)
    transform = src.transform

colores = ["green", "red"]
cmap = ListedColormap(colores)

plt.figure(figsize=(8, 10))
plt.imshow(CentroSldVul_reclass, cmap=cmap, extent=(transform[2], transform[2] + transform[0]*CentroSldVul_reclass.shape[1],
                                            transform[5] + transform[4]*CentroSldVul_reclass.shape[0], transform[5]))


cbar = plt.colorbar(ticks=[1, 3])
cbar.ax.set_yticklabels(["Baja", "Alta"])
cbar.set_label("Categoría de vulnerabilidad")

plt.title("Vulnerabilidad de PM2.5 de acuerdo a los centros de salud")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(False)
plt.savefig(f"/notebooks/Proyecto/Calidad_Aire_Soacha/Resultados/VulnerabilidadCM_PM25.png", dpi=300)
plt.show()
```
